```python
# Dependencies
import pandas as pd

# File to Load & Read
school_csv = pd.read_csv("../Resources/schools_complete.csv")
student_csv = pd.read_csv("../Resources/students_complete.csv")

# Combine the data into a single dataset
school_student_data = pd.merge(school_csv, student_csv, how="left", on=["school_name", "school_name"])
school_student_data.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(school_student_data['school_name'].unique())
#print(total_schools)

total_students = len(school_student_data['student_name'])
#print('{:,}'.format(total_students))

total_budget = (school_student_data['budget'].unique()).sum()
#print('${:,}'.format(total_budget))

average_math = school_student_data['math_score'].mean()
#print('{:,.6f}'.format(average_math))

average_reading = school_student_data['reading_score'].mean()
#print('{:,.6f}'.format(average_reading))

overall_average = (average_math + average_reading)/2
#print('{:,.6f}'.format(overall_average))

passing_math = (((school_student_data ['math_score'] >= 70).sum())/ (len(school_student_data ['math_score'] >= 70))*100)
#print('{:,.6f}'.format(passing_math))

passing_reading = (((school_student_data ['reading_score'] >= 70).sum())/ (len(school_student_data ['reading_score'] >= 70))*100)
#print('{:,.6f}'.format(passing_reading))

overall_passing = (passing_math + passing_reading)/2
#print('{:,.6f}'.format(overall_passing))

district_summary = pd.DataFrame({"Total School": [total_schools], 
                                 "Total Students": ['{:,}'.format(total_students)], 
                                 "Total Budget": ['${:,}'.format(total_budget)], 
                                 "Average Math Score": ['{:,.6f}'.format(average_math)], 
                                 "Average Reading Score": ['{:,.6f}'.format(average_reading)], 
                                 "Overall Average Score": ['{:,.6f}'.format(overall_average)], 
                                 "% of Students w/ Passing Math Scores": ['{:,.6f}'.format(passing_math)], 
                                 "% of Students w/ Passing Reading Scores": ['{:,.6f}'.format(passing_reading)], 
                                 "Overall Passing Rate": ['{:,.6f}'.format(overall_passing)]
                                })
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_name = school_student_data.groupby('school_name')
#school_name.head()

school_type = school_name['type'].first()
#school_type

school_students = school_name['student_name'].count()
#school_students

school_budget = school_name['budget'].first()
#school_budget

per_student_budget = school_budget / school_students
#per_student_budget

avg_math_by_school = school_name['math_score'].mean()
#avg_math_by_school

avg_reading_by_school = school_name['reading_score'].mean()
#avg_reading_by_school

all_passing_math = school_student_data[school_student_data ['math_score']>=70].groupby(['school_name']).size()
passing_math_students = (all_passing_math/school_students)*100
#passing_math_students

all_passing_reading = school_student_data[school_student_data ['reading_score']>=70].groupby(['school_name']).size()
passing_reading_students = (all_passing_reading/school_students)*100
#passing_reading_students

overall_passing_schools = (passing_math_students + passing_reading_students)/2
#overall_passing_schools

school_data = {'School Type': school_type, 
                  'Total Students': school_students,
                  'Total School Budget': school_budget,
                  'Per Student Budget': per_student_budget,
                  'Average Math Score': avg_math_by_school,
                  'Average Reading Score': avg_reading_by_school,
                  '% Passing Math': passing_math_students,
                  '% Passing Reading': passing_reading_students,
                  '% Overall Passing Rate': overall_passing_schools
                 }
    
school_summary = pd.DataFrame (school_data)
school_summary.head()

final_school_summary = school_summary.copy()
final_school_summary ['Total Students'] = final_school_summary['Total Students'].map('{:,}'.format)
final_school_summary['Per Student Budget'] = final_school_summary['Per Student Budget'].map('${:,.2f}'.format)
final_school_summary['Total School Budget'] = final_school_summary['Total School Budget'].map('${:,.2f}'.format)
final_school_summary.index.name = None

final_school_summary.head()
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
top_schools = final_school_summary.sort_values(by='% Overall Passing Rate',ascending=False)
top_schools.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
bottom_schools = final_school_summary.sort_values(by='% Overall Passing Rate',ascending=True)
bottom_schools.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
math_9th_avg = school_student_data[school_student_data['grade']=='9th'].groupby('school_name')['math_score'].mean()
#math_9th_avg

math_10th_avg = school_student_data[school_student_data['grade']=='10th'].groupby('school_name')['math_score'].mean()
#math_10th_avg

math_11th_avg = school_student_data[school_student_data['grade']=='11th'].groupby('school_name')['math_score'].mean()
#math_11th_avg

math_12th_avg = school_student_data[school_student_data['grade']=='12th'].groupby('school_name')['math_score'].mean()
#math_12th_avg

all_math_scores = {'9th': math_9th_avg,
                   '10th': math_10th_avg,
                   '11th': math_11th_avg,
                   '12th': math_12th_avg
                  }

math_score_by_grade = pd.DataFrame(all_math_scores)
math_score_by_grade.index.name = None

math_score_by_grade
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
reading_9th_avg = school_student_data[school_student_data['grade']=='9th'].groupby('school_name')['reading_score'].mean()
#reading_9th_avg

reading_10th_avg = school_student_data[school_student_data['grade']=='10th'].groupby('school_name')['reading_score'].mean()
#reading_10th_avg

reading_11th_avg = school_student_data[school_student_data['grade']=='11th'].groupby('school_name')['reading_score'].mean()
#reading_11th_avg

reading_12th_avg = school_student_data[school_student_data['grade']=='12th'].groupby('school_name')['reading_score'].mean()
#reading_12th_avg

all_reading_scores = {'9th': reading_9th_avg,
                   '10th': reading_10th_avg,
                   '11th': reading_11th_avg,
                   '12th': reading_12th_avg
                  }

reading_score_by_grade = pd.DataFrame(all_reading_scores)
reading_score_by_grade.index.name = None

reading_score_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
scores_by_school_spending = school_summary.loc[:,['Average Math Score',
                                  'Average Reading Score','% Passing Math',
                                  '% Passing Reading','% Overall Passing Rate']]

scores_by_school_spending['Spending Ranges (Per Student)']= pd.cut(school_summary['Per Student Budget'], 
                                                                   spending_bins, labels=group_names)

scores_by_school_spending = scores_by_school_spending.groupby('Spending Ranges (Per Student)').mean()

scores_by_school_spending
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Sample bins. Feel free to create your own bins.
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
scores_by_school_size = school_summary.loc[:,['Average Math Score',
                                  'Average Reading Score','% Passing Math',
                                  '% Passing Reading','% Overall Passing Rate']]

scores_by_school_size['School Size']= pd.cut(school_summary['Total Students'],
                                             size_bins, labels=group_names)

scores_by_school_size = scores_by_school_size.groupby('School Size').mean()

scores_by_school_size
```

## Scores by School Type

* Perform the same operations as above, based on school type.

```python
scores_by_school_type = school_summary[['School Type','Average Math Score',
                                  'Average Reading Score','% Passing Math',
                                  '% Passing Reading','% Overall Passing Rate',]]

scores_by_school_type = scores_by_school_type.groupby('School Type').mean()
scores_by_school_type.head()
```

**Observations**

-  Both math and reading average scores for each school do not sustain significant changes when comparing students across all high school grade levels.
